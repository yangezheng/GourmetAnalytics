```python
import pandas as pd
import numpy as np
```

```python
file1 = '../Data/diet.csv'
file2 = '../Data/recipes.csv'
file3 = '../Data/requests.csv'
file4 = '../Data/reviews.csv'

diet = pd.read_csv(file1)
recipes = pd.read_csv(file2)
requests = pd.read_csv(file3)
reviews = pd.read_csv(file4)
```

## Make some groudn truth

```python
diet_reviews  = pd.merge(diet,reviews,how='outer', on= 'AuthorId')
diet_reviews_recipes = pd.merge(diet_reviews,recipes,how='outer',on='RecipeId')
# diet_reviews_recipes_requests = pd.merge(diet_reviews_recipes,requests,how='outer', on=['AuthorId','RecipeId'])
df = diet_reviews_recipes
```

```python
df.info()
```

```python
non_numeric_cols = df.select_dtypes(include=['object']).columns
non_numeric_cols
```

```python
df['Like'] = df['Like'].astype('category')
# df['RecipeCategory'] = df['RecipeCategory'].astype('category')
# df['Diet'] = df['Diet'].astype('category')
df = df.drop(columns=['AuthorId', 'Name','TestSetId' ],errors='ignore')
df = df.dropna(subset=['Like'])
```

```python
# delete because I don't know how to use them
df = df.drop(columns=['RecipeIngredientQuantities', 'RecipeIngredientParts','RecipeYield'],errors='ignore')
```

## Random Forest

```python
clean_df = df.dropna()
```

```python
from sklearn.ensemble import RandomForestClassifier

train_model = RandomForestClassifier(n_estimators=5, max_features=3, random_state=2023+2024)

X = clean_df.drop(columns=["Like"])
y = clean_df["Like"]

train_model.fit(X,y)
```

## Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
pred = train_model.predict(X)
error_rate = np.mean(y != pred)
print("Error rate:", error_rate)
print("Accuracy:", accuracy_score(y, pred))
```
